```python
def bisection(xl,xu,tol,max_iteration):
  iteration=1
  xr_old=None
  results=[]
  while iteration<max_iteration:
    xm=(xl+xu)/2
    fxm=function(xm)
    if xr_old is not None:
      relative_error = np.abs((xm - xr_old) / xm) * 100;
    else:
      relative_error = None
    results.append([iteration,xl,xu,xm,fxm,relative_error])
    if fxm == 0.0 or (relative_error is not None and relative_error < tol):
            break
    if np.sign(function(xl)) == np.sign(fxm):
            xl = xm
    else:
            xu = xm
    xr_old = xm
    iteration += 1
  df=pd.DataFrame(results,columns=['iteration','xl','xu','xm','fxm','relative_error'])
  return df
```

```python
# Input values for the problem
xl = 1.0            # Lower bound
xu = 2.0           # Upper bound
tolerance = 0.01    # Error tolerance in percentage
max_iterations = 50 # Maximum number of iterations

# Run the bisection method
bisection(xl, xu, tolerance, max_iterations)
```

```python
import numpy as np
import pandas as pd

# Define the function and its derivative
def function(x):
    # Example: f(x) = x^2 - 2
    return x**2 - 2

def deri(x):
    # Example: derivative of f(x) = 2x
    return 2*x

def newton(x, tol, max_iteration):
    iteration = 0
    results = []

    while iteration < max_iteration:
        # Compute next approximation
        xn = x - function(x) / deri(x)

        # Calculate the function value at the new point
        funct = function(xn)

        # Calculate relative error
        relative_error = np.abs((xn - x) / xn) * 100

        # Append results for the current iteration
        results.append([iteration, x, xn, funct, relative_error])

        # Check if the relative error is within tolerance
        if relative_error < tol:
            break

        # Update the current value and increase iteration count
        iteration += 1
        x = xn

    # Convert results to a pandas DataFrame
    df = pd.DataFrame(results, columns=['iteration', 'x', 'xn', 'funct', 'relative_error'])

    return df

# Example usage:
x0 = 1.0  # Initial guess
tolerance = 1e-6  # Tolerance for stopping criterion
max_iter = 100  # Maximum number of iterations

df_results = newton(x0, tolerance, max_iter)
print(df_results)
```
