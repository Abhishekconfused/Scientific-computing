```python
import numpy as np#libraries that make life easy
import pandas as pd
```

```python
pd.set_option('display.float_format', '{:.15f}'.format)# setting the decimal places upto which I want to print
```

```python
def function(x):
  return (x*x)-2#Function given in the question
```

```python
def derivative(x):
  return 2*x#Derivative of the function
```

```python
def newton_Raphson(x,tolerated_error,max_iterations):
  iteration=0#starting indexing at 0
  x_new=None#where I am storing my Xn+1
  results = []#Empty dictionery for results
  while iteration < max_iterations:#while loop where I am putting max_iteration condition
    x_new = x - (function(x) / derivative(x))
    funct = function(x_new)#function at new value
    deri=derivative(x_new)#derivative at new value
    relative_error = np.abs((x_new - x) / x_new) # error calculation
    results.append([iteration, x, x_new, funct,deri, relative_error])#result storing
    if funct == 0.0 or (relative_error is not None and relative_error < tolerated_error):#condition to stop
            break
    x = x_new# updating the x value
    iteration += 1# updating the iteration value
  df=pd.DataFrame(results,columns=['Iteration','Xn','Xn+1','f(Xn+1)','derivative_at_x_new','Relative Error'])#storing in tabular form,Sir told I need to make tables for final result so this thing written
  return df

```

```python
x=2
tolerated=0.0000001
max_iterations=100
newton_Raphson(x,tolerated,max_iterations)
```
